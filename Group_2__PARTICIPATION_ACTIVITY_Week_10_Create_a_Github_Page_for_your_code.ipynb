```python
import numpy as np
import pandas as pd
import networkx as nx
from matplotlib import pyplot as plt
import random
from numpy import random as nprand
```

```python
students = pd.read_csv("/content/drive/MyDrive/MSSPDA 6070/Week_10/students.csv")
students
```

```python
connection = pd.read_csv("/content/drive/MyDrive/MSSPDA 6070/Week_10/connection.csv")
connection
```

```python
G = nx.Graph()

for index, row in students.iterrows():
    G.add_node(row['Name'], Age=row['Age'], Gender=row['Gender'], Major=row['Major'], Interests=row['Interests'])

for index, row in connection.iterrows():
    G.add_edge(row['Student_ID_1'], row['Student_ID_2'])
```

```python
import matplotlib.pyplot as plt
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 10))
nx.draw(G, pos, with_labels=True, node_color='skyblue', edge_color='gray', node_size=3000)
plt.show()
```
